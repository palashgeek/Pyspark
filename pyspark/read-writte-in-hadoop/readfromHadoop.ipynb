```python
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder \
    .appName("Read Files from Hadoop") \
    .getOrCreate()

# Read JSON file from Hadoop
json_df = spark.read.option("mode", "PERMISSIVE").json("hdfs://localhost:9000/Hadoop_File/jsonfile")

# Read CSV file from Hadoop
csv_df = spark.read.csv("hdfs://localhost:9000/Hadoop_File/csvfile")

# Read Parquet file from Hadoop
parquet_df = spark.read.parquet("hdfs://localhost:9000/Hadoop_File/parquet")

# Show the dataframe schema and some sample data
json_df.printSchema()
'''json_df.select("_corroupt_record").show()'''

csv_df.printSchema()
csv_df.show()

parquet_df.printSchema()
parquet_df.show()
```

```python
from pyspark.sql import SparkSession

# Create SparkSession
spark = SparkSession.builder \
    .appName("Read CSV and Write JSON") \
    .getOrCreate()

# Read CSV file
csv_df = spark.read.csv("hdfs://localhost:9000/Hadoop_File/csvfile", header=True, inferSchema=True)

# Show the dataframe schema and some sample data
csv_df.printSchema()
csv_df.show(truncate=False)

# Write DataFrame to JSON file
csv_df.write.json("hdfs://localhost:9000/Hadoop_File/csv_to_json_output")

# Stop SparkSession
spark.stop()
```
